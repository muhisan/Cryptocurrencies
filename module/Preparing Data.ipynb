```python
file_path = "Resources/shopping_data.csv"
df_shopping = pd.read_csv(file_path,encoding="ISO-8859-1")
df_shopping.head(5)
```

```python
# Columns
df_shopping.columns
```

```python
#List dataframe data types
df_shopping.dtypes
```

```python
#Find null values
for column in df_shopping.columns:
    print(f"Column {column} has {df_shopping[column].isnull().sum()} null values")
```

```python
# Drop null rows
df_shopping = df_shopping.dropna()
```

```python
#Find duplicate entries
print(f"Duplicate entries: {df_shopping.duplicated().sum()}")
```

```python
#Remove the CustomerID column
df_shopping.drop(columns=["CustomerID"],inplace=True)
df_shopping.head()
```

```python
#Transform String column
def change_string(member):
    if member == 'Yes':
        return 1
    else:
        return 0
    
df_shopping["Card Member"]= df_shopping["Card Member"].apply(change_string)
df_shopping.head()
```

```python
#Transform annual income
df_shopping["Annual Income"]= df_shopping["Annual Income"]/1000
df_shopping.head()
```

```python
df_shopping.rename(columns={"Spending Score (1-100)": "Spending Score"}, inplace=True)
df_shopping.head()
```

```python
df_shopping.rename(columns={"Card Member": "Genre"}, inplace=True)
df_shopping.head()
```

```python
df_shopping.columns = df_shopping. columns. str. replace(' ',"")
df_shopping.head()
```

```python
#df['values'].round(decimals = 3)
m=(df_shopping.dtypes=='float')
df_shopping.loc[:,m]=df_shopping.loc[:,m].astype(int)
df_shopping.head()
```

```python
# Saving cleaned data
file_path = "Resources/shopping_data_cleaned.csv"
df_shopping.to_csv(file_path, index=False)
```
